### Handling outliers

```python
num_cols =df.select_dtypes(include='number').columns
num_cols
```

```python
from scipy import stats
iqr = stats.iqr(df['Price'])
iqr
```

```python
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3-q1
iqr
```

```python
upper_bound = q1-1.5*iqr
lower_bound =q3+1.5*iqr
print(upper_bound)
print(lower_bound)
```

```python
df = df[(df['Price']>= upper_bound)&(df['Price']<=lower_bound)]
```

```python
df
```

### extracting date_of _Journy 

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
df['Date_of_Journey']
```

```python
df['Day_of_Journey']=df['Date_of_Journey'].dt.day
df['Day_of_Journey']
```

```python
df['Month_of_journey']= df['Date_of_Journey'].dt.month
df['Month_of_journey']
```

```python
df['Dep_Hour']=pd.to_datetime(df['Dep_Time']).dt.hour
df['Dep_Minute']=pd.to_datetime(df['Dep_Time']).dt.minute
```

```python
df['Arrival_Hour']=pd.to_datetime(df['Arrival_Time']).dt.hour
df['Arrival_Minute']=pd.to_datetime(df['Arrival_Time']).dt.minute
```

```python
df[['Arrival_Hour','Arrival_Minute']]
```

```python
df
```

```python
def convert_duration(duration):
    hours,minutes = 0,0
    if 'h' in duration:
        hours = int(duration.split('h')[0])
    if 'm'in duration:
        minutes = int(duration.split('h')[-1].split('m')[0])
    return hours*60+minutes 

df['Duration_Minutes']=df['Duration'].apply(convert_duration)
df['Duration_Minutes']
```

```python
df
```

```python
df.drop(['Date_of_Journey','Dep_Time','Arrival_Time','Duration'],axis=1,inplace=True)
```

## Model creation

### data spliting

```python
x = df.drop('Price',axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

### Random forest

```python
ransom_forest_model = RandomForestRegressor()
ransom_forest_model.fit(x_train,y_train)
```

```python
y_pred = ransom_forest_model.predict(x_test)
```

```python
ransom_forest_mae = mean_absolute_error(y_test,y_pred)
ransom_forest_mse = mean_squared_error(y_test,y_pred)
ransom_forest_r2  = r2_score(y_test,y_pred)
print(f'Mean Absolute error = {ransom_forest_mae}')
print(f'Mean squared error = {ransom_forest_mse}')
print(f'r2_score ={ransom_forest_r2}')
```

### Hyper parameter tuning for random forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the parameter grid
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the Grid Search
grid_search_rf = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid_rf, cv=3, n_jobs=-1, verbose=2)

# Fit the Grid Search to the data
grid_search_rf.fit(x_train, y_train)

# Get the best parameters
best_params_rf = grid_search_rf.best_params_
print(f'Best Parameters for Random Forest: {best_params_rf}')
```

```python
best_params_rf={'max_depth': 30,
                'min_samples_leaf': 1,
                'min_samples_split': 2,
                'n_estimators': 200
               }
model_rf = RandomForestRegressor(**best_params_rf,random_state=42)
```

```python
model_rf.fit(x_train,y_train)
```

```python
model_rf_pred = model_rf.predict(x_test)
```

```python
model_rf_mae = mean_absolute_error(y_test,model_rf_pred)
model_rf_mse = mean_squared_error(y_test,model_rf_pred)
model_rf_r2  = r2_score(y_test,model_rf_pred)
print(f'Mean Absolute error = {model_rf_mae}')
print(f'Mean squared error = {model_rf_mse}')
print(f'r2_score ={model_rf_r2}')
```

### Hyper parameter tuning for gradient boosting

```python
param_grid_gb = {
    'n_estimators':[100,200,300],
    'learning_rate':[0.01,0.1,0.2],
    'max_depth':[3,4,5],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
grid_search_gb = GridSearchCV(estimator=GradientBoostingRegressor(),param_grid=param_grid_gb,n_jobs=-1,verbose=2)
grid_search_gb.fit(x_train,y_train)
best_params_gb = grid_search_gb.best_params_
print(f'best params for gb tuning={best_params_gb}')
```

```python
best_model_gb = GradientBoostingRegressor(**best_params_gb)
best_model_gb.fit(x_train,y_train)
```

```python
best_model_gb_pred = best_model_gb.predict(x_test)
```

```python
best_model_gb_mae = mean_absolute_error(y_test,best_model_gb_pred)
best_model_gb_mse = mean_squared_error(y_test,best_model_gb_pred)
best_model_gb_r2 = r2_score(y_test,best_model_gb_pred)
print(f'Mean absolute error = {best_model_gb_mae}')
print(f'mean squared error = {best_model_gb_mse}')
print(f'r2_score= {best_model_gb_r2}')
```
