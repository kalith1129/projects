```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Cricket (1).csv',encoding='latin1')
df# the encoding parameter to ensure that the file is read correctly without errors, especially when dealing with non-ASCII characters
# Player: Name and country of the cricketer.
# Span: Years active in international cricket.
# Mat: Total matches played.
# Inns: Total innings batted.
# NO: Number of times not out.
# Runs: Total runs scored.
# HS: Highest score in an innings.
# Ave: Batting average (runs per innings).
# BF: Balls faced.
# SR: Strike rate (runs per 100 balls).
# 100/50: Centuries (100+) and half-centuries (50-99).
# 0: Number of times dismissed for zero runs.
```

```python
df.drop('Player',axis=1,inplace=True)
```

```python
df.info()
```

```python
df[['start','end']]=df['Span'].str.split('-',expand=True)
df
```

```python
df.info()
```

```python
df['start']=df['start'].astype('int')
df['end']=df['end'].astype('int')
```

```python
df['exp']=df['end']-df['start']
df
```

```python
df.drop('Span',axis=1,inplace=True)
```

```python
df['HS']=df['HS'].str.extract('(\d+)')#regular expression (regex) used to find and extract numeric values from strings
#D+-->extract abc from abc123
df
```

```python
df['HS']=df['HS'].astype('int')
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap='rainbow')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
df_scale=scale.fit_transform(df)
df_scale
```

```python
df2=pd.DataFrame(data=df_scale,columns=df.columns)
df2
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,12):
    #kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,12),wcss,marker='o',linestyle='-',color='b')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(np.linspace(0,11, 12, dtype=int))
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42)

#Training the model
model.fit(df2)
```

```python
label=model.labels_
label
```
